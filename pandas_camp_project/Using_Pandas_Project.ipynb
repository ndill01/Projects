```python
import pandas as pd
import numpy as np
```

```python
# Setting up a dataframe that takes information from the dump_application.csv file.
application_df = pd.read_csv('dump_application.csv').dropna(axis = 1, how = 'all')

# Take the shape of application_df to find the amount of campers that have applied.
application_df.shape
```

```python
# Setting up a dataframe that takes information from the dump_registration.csv file.
registration_df = pd.read_csv('dump_registration.csv').dropna(axis = 1, how = 'all')

# Take the shape of registration_df to find the amount of registered campers.
registration_df.shape
```

```python
# Setting up a dataframe that takes information from the dump_camp.csv file.
camp_df = pd.read_csv('dump_camp.csv').dropna(axis = 1, how = 'all')

# Using nested for loops to find the amount of campers that have attended each camp.
for x in camp_df['id']:
    count = 0
    print(x)
    for y in application_df['camp_id']:
        if x == y:
            count += 1
    print(count)
    print()
```

```python
# Creating a list of the different t-shirt sizes.
t_shirt_list = ['Adult 2XL', 'adult_xxl', 'Adult Extra Large', 'adult_xl', 'Adult Large', 'adult_large', 'Adult Medium', 'adult_meduim', 
                    'Adult Small', 'adult_small', 'Child Large', 'child_large', 'Child Medium', 'child_medium', 'Child Small', 'child_small']

# Using nested for loops to find the amount of campers per t-shirt size.
for x in t_shirt_list: 
    count = 0
    print(x)
    for y in registration_df['t_shirt_size']:
        if y == x:
            count += 1
    print(count)
    print()
```

```python
# Making a list of the different season identifiers.
season_list = [1, 2, 3, 4, 5, 6, 7]

# Making a dataframe containing prices above 0.
season_price_df = camp_df[camp_df['price'] > 0]

# Using a nested for loop to total up the price for each camp season.
for x in season_list:
    print(x)
    price_total = 0
    count = 0
    for y in season_price_df['season_id']:
        if y == x:
            price_total += season_price_df['price'].iloc[count]
        count += 1
    print(price_total)
    print()
```

```python
# Setting up a dataframe that takes information from the dump_account.csv file.
account_df = pd.read_csv('dump_account.csv').dropna(axis = 1, how = 'all')

# Creating an empty list that will be used to store the zipcodes of the campers.
zipcode_list = []

# Using a for loop to add each zipcode to the empty list.
for x in account_df['zipcode']:
    zipcode_list.append(x)
    
# Making a new dictionary that will contain the zipcodes and their occurrences.
zipcode_dictionary = dict()

# Using a for loop with an if-else statement to check if the zipcode appears multiple times and increment its occurrences.
for y in zipcode_list:
    if y in zipcode_dictionary:
        zipcode_dictionary[y] += 1
    else:
        zipcode_dictionary[y] = 1

# A filter for the pairs between the zipcode number and its total amount of occurrences to only keep zipcodes with multiple occurrences.      
zipcode_dictionary = {zipcode:campers for zipcode, campers in zipcode_dictionary.items() if campers > 1}

# A for loop that prints each zipcode and the amount of campers to the screen.
for zipcode, campers in zipcode_dictionary.items():
    print(zipcode, ' ', campers)
```
